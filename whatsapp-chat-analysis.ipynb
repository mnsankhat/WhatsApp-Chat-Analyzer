```python
import re
import pandas as pd 
```

```python
f=open("WhatsApp Chat with STUDY MATERIAL GROUP✍️.txt",'r',encoding = 'utf-8')
```

```python
data = f.read()
```

```python
print(data)
```

```python
print(type(data))
```

```python
pattern = r"\d{1,2}/\d{1,2}/\d{2},\s\d{1,2}:\d{2}\s[AP]M\s-\s"
```

```python
messages = re.split(pattern,data)[1: ]
messages
```

```python
dates = re.findall(pattern,data)
dates
```

```python
df = pd.DataFrame({'user_message':messages, "message_date":dates})
# conver message_data types
df['message_date'] = df['message_date'].str.replace('\u202f', ' ', regex=False)

df['message_date'] = pd.to_datetime(
    df['message_date'],
    format='%d/%m/%y, %I:%M %p - '
)

df.rename(columns={'message_date': 'date'}, inplace=True)

df.head()
```

```python

users = []
messages = []
for message in df['user_message']:
    entry = re.split('([\w\W]+?):\s', message)
    if entry[1:]:  # user name
        users.append(entry[1])
        messages.append(" ".join(entry[2:]))
    else:
        users.append('group_notification')
        messages.append(entry[0])

df['user'] = users
df['message'] = messages
df.drop(columns=['user_message'], inplace=True)

df.head(10)
```
